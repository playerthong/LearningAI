# Import libs 

```python
import pandas as pd
from pathlib import Path
```

# Import data

```python
current_folder = Path.cwd()
print("Current folder:", current_folder)

train_file_path = str(current_folder) + '/data/train.csv'
print("Train file:", train_file_path)
train_data = pd.read_csv(train_file_path) 
train_data.columns
```

```python
test_file_path = str(current_folder) + '/data/test.csv'
print("Test file:", test_file_path)
test_data = pd.read_csv(test_file_path) 
test_data.columns
```

# Target prediction

```python
y = train_data.SalePrice
train_data.info()
```

```python
house_features = [ 'BedroomAbvGr', 'GarageArea']
X_train = train_data[house_features]
X_train.describe()
```

```python
X_train.head()
```

```python

X_test = test_data[house_features]
X_test.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor
# Define model. Specify a number for random_state to ensure same results each run
house_model = DecisionTreeRegressor(random_state=1)

# Fit model
house_model.fit(X_train, y)
```

# Predict

```python
print("Making predictions for the following 5 houses:")
print(X_test.head())
print("The predictions are")
print(house_model.predict(X_test.head()))
```
